```python
import warnings
warnings.filterwarnings('ignore')
import os
import anthropic
import fitz
from google.colab import userdata, files

# Setup
client = anthropic.Anthropic(
    api_key=userdata.get("ANTHROPIC_API_KEY")
)

# ============================================
# PDF READING FUNCTION
# ============================================
def read_pdf(pdf_path):
    """Extract text from PDF file"""
    doc = fitz.open(pdf_path)
    text = ""
    for page in doc:
        text += page.get_text()
    doc.close()
    return text

# ============================================
# PDF Q&A AGENT
# ============================================
def pdf_qa_agent(pdf_text, question):
    """Answer questions based on PDF content"""
    context = pdf_text[:8000]

    response = client.messages.create(
        model="claude-haiku-4-5",
        max_tokens=1024,
        system="You are a PDF analysis expert. Answer questions based only on the provided PDF content. Be precise and cite relevant parts. Use plain text only.",
        messages=[{
            "role": "user",
            "content": f"PDF Content:\n{context}\n\nQuestion: {question}"
        }]
    )
    return response.content[0].text

# ============================================
# DOWNLOAD SAMPLE PDF
# ============================================
print("Downloading sample research paper...")
import subprocess
subprocess.run([
    "wget", "-q", "-O", "paper.pdf",
    "https://arxiv.org/pdf/1706.03762.pdf"
])
print("Downloaded: Attention Is All You Need paper!")

# ============================================
# TEST THE AGENT
# ============================================
print("\nReading PDF...")
pdf_text = read_pdf("paper.pdf")
print(f"Characters read: {len(pdf_text)}")

print("\n" + "=" * 50)
print("PDF Q&A Agent Ready!")
print("=" * 50)

print("\nQuestion 1:")
print(pdf_qa_agent(pdf_text, "What is this paper about?"))

print("\nQuestion 2:")
print(pdf_qa_agent(pdf_text, "What is the attention mechanism?"))

print("\nQuestion 3:")
print(pdf_qa_agent(pdf_text, "Who are the authors?"))
```
